```python
## convert phenosys , to . & ; to ,

import pandas as pd
import chardet
import matplotlib.pyplot as plt
import os
import glob

Phenosys_folder= '/Users/hunglo/Documents/inscopix_csv/phenosys_test/'
extension = 'csv'
os.chdir(Phenosys_folder)
result = sorted(glob.glob('*.{}'.format(extension)))

for r in result[:]:
    print(r)
    file_path = os.path.join(Phenosys_folder,r)
    # file_path = '/Users/hunglo/Documents/inscopix_csv/phenosys_test/DSC018624_Inscopix-23.01.19.csv'

    with open(file_path, 'rb') as f:
        result = chardet.detect(f.read())  # or readline if the file is large

    # print(result)
    df = pd.read_csv(file_path, encoding=result['encoding'],on_bad_lines='skip',sep=';')

    output_folder = '/Users/hunglo/Documents/inscopix_csv/phenosys_test/converted/'

    df.to_csv(f'{os.path.join(output_folder,r[:-4])}_.csv',index=False)

    file_path_converted = f'{os.path.join(output_folder,r[:-4])}_.csv'
    print('done')
print('All done!')
    # import sys
    # sys.path.append('/Users/hunglo/Documents/python_codes/')
    # import binge

    # L1,P1,P2,PAll = binge.load_phenosys(file_path_converted)
    # plt.figure(figsize=[12,4])
    # plt.eventplot([L1,P1,P2,PAll],linelengths=0.5,linewidths=0.5)
    # plt.show()
```
